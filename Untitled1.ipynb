```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Aman", "Riya"],
    "Marks": [85, 90, 78, 95, 88]
}

df = pd.DataFrame(data)

print(df.describe())
```

```python
import matplotlib.pyplot as plt

plt.pie(df["Marks"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Student Marks Distribution")
plt.show()
```

```python
plt.plot(df["Name"], df["Marks"], marker="o")
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Aman", "Riya"],
    "Marks": [85, None, 78, 95, None]
}

df = pd.DataFrame(data)

print(df)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print(df)
```

```python
sorted_df = df.sort_values(by="Marks", ascending=False)

print(sorted_df)
```

```python
top_student = df.loc[df["Marks"].idxmax()]

print(top_student)
```

```python
top3 = df.nlargest(3, "Marks")

print(top3)
```

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 80 else "Fail")

print(df)
```

```python
df.to_csv("students_result.csv", index=False)

print("CSV file saved successfully!")
```
